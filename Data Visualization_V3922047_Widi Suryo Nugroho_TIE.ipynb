```python
import pandas as pd
import matplotlib.pyplot as plt

# reading the database
data = pd.read_csv("salesdata.csv")

# Scatter plot with day againts tip
plt.scatter(data['Category'], data['Quantity'])

# Adding title to the plot
plt.xlabel('Category')
plt.ylabel('Quantity')



plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("salesdata.csv")

plt.plot(data['Category'])
plt.plot(data['Quantity'])

plt.title("Data Sales")

plt.xlabel('Category')
plt.ylabel('Quantity')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("salesdata.csv")

plt.bar(data['Category'], data['Quantity'])


plt.title("Data Sales Bar Chart")

plt.xlabel('Category')
plt.ylabel('Quantity')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("salesdata.csv")

plt.hist(data['Profit'])

plt.title("Profit Data Sales Histogram")

plt.show()
```
